```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the sales data
file_path = "sales_data.csv"  # Make sure the file is uploaded in Colab
df = pd.read_csv(file_path)

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Set Seaborn style
sns.set_style("whitegrid")
```

```python
def plot_sales_trend():
    """Plots sales trend over time."""
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=df, x="Date", y="UnitsSold", hue="Product", marker="o")
    plt.title("Sales Trend Over Time")
    plt.xlabel("Date")
    plt.ylabel("Units Sold")
    plt.xticks(rotation=45)
    plt.legend(title="Product")
    plt.show()

plot_sales_trend()  # Automatically plot
```

```python
def plot_units_sold():
    """Plots total units sold per product."""
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df, x="Product", y="UnitsSold", estimator=sum, ci=None, palette="viridis")
    plt.title("Total Units Sold per Product")
    plt.xlabel("Product")
    plt.ylabel("Total Units Sold")
    plt.show()

plot_units_sold()  # Automatically plot
```

```python
def plot_revenue_distribution():
    """Plots revenue distribution among products."""
    revenue_per_product = df.groupby("Product")["Revenue"].sum()
    plt.figure(figsize=(7, 7))
    plt.pie(revenue_per_product, labels=revenue_per_product.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
    plt.title("Revenue Distribution Among Products")
    plt.show()

plot_revenue_distribution()  # Automatically plot
```
